## Análise de Normalidade

Teste Shapiro-Wilk <br>

H0 - distribuição normal: p > 0.05 <br>
Ha - distribuição não-normal: p <= 0.05

```python
import scipy.stats as stats
```

```python
# Nota de redação
stats.shapiro(notas_colegiox.NOTA_REDACAO)
# > 0.05 = normal
```

```python
# Nota final
stats.shapiro(notas_colegiox.nota_final)
```

```python
# Nota de matemática
stats.shapiro(notas_colegiox.NOTA_MT)
```

```python
# Nota de ciências da natureza
stats.shapiro(notas_colegiox.NOTA_CN)
```

```python
# Nota de linguagens e códigos
stats.shapiro(notas_colegiox.NOTA_LC)
```

```python
# Nota de ciências humanas
stats.shapiro(notas_colegiox.NOTA_CH)
# Não é normal
```

```python
stats.probplot(notas_colegiox['NOTA_REDACAO'], dist="norm", plot=plt)
plt.title('Normal Q-Q plot')
plt.show()
```

```python
stats.probplot(notas_colegiox['NOTA_CH'], dist="norm", plot=plt)
plt.title('Normal Q-Q plot')
plt.show()
```

## Regressão

#### Regressão linear com statsmodels

```python
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
```

```python
regressao = smf.ols('nota_final ~ NOTA_REDACAO', data=notas_colegiox).fit()
print(regressao.summary())
```

```python
coef = pd.DataFrame(regressao.params)
coef.columns = ['Coeficientes']
print(coef)
```

#### Equação:
Notal final = 251.678855 + (0.362749 X Nota Redação)

```python
regressao.predict()
```

```python
residuos = regressao.resid
residuos
```

```python
plt.scatter(y=notas_colegiox.nota_final,
           x=notas_colegiox.NOTA_REDACAO, color="blue", s=60, alpha=.4)
#X_plot = np.linspace(min(notas_colegiox.NOTA_REDACAO), max(notas_colegiox.NOTA_REDACAO), len(notas_colegiox.NOTA_REDACAO))
X_plot = np.linspace(0, 1000)
plt.plot(X_plot, X_plot*regressao.params[1] + regressao.params[0], color='r')
plt.title('Reta de regressão')
plt.ylabel('Notal final')
plt.xlabel('Nota de redação')
plt.show()
```

#### Essa reta obedece a equação 
Notal final = 251.678855 + (0.362749 X Nota Redação)

### Teste de normalidade dos resíduos

```python
stats.shapiro(residuos)
```

```python
stats.probplot(residuos, dist='norm', plot=plt)
plt.title('Normal Q-Q plot - Resíduos')
plt.show()
```

## Análise de Homocedasticidade dos resíduos
(resíduos com variação constante)

```python
plt.scatter(y=residuos, x=regressao.predict(), color='blue')
plt.hlines(y=0,xmin=320, xmax=530, color='orange')
plt.ylabel('Resíduos')
plt.xlabel('Valores preditos')
plt.show()
```

## Regressão com Sklearn

```python
x = notas_colegiox.iloc[:, 4].values
y = notas_colegiox.iloc[:, 5].values
```

```python
# Nota redação, nota final
correlacao2 = np.corrcoef(x, y)
correlacao2
```

```python
# Transformando em matriz
x = x.reshape(-1, 1) 
```

```python
from sklearn.linear_model import LinearRegression
regressao2 = LinearRegression()
regressao2.fit(x, y)
```

```python
regressao2.intercept_
```

```python
# Coeficiente angular
regressao2.coef_
```

```python
# Coeficiente de determinação
regressao2.score(x, y)
```

```python
#
previsoes = regressao2.predict(x)
previsoes
```

```python
# Se a notas da redação for o valor entre [ ], qual será a nota final?
# Previsão com base na reta de regressão
previsao = regressao2.predict([[330]])
previsao
```

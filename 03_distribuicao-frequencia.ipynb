# Distribuição de frequências

```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
enem_sp = pd.read_csv('data/enem_2019-tratado.csv', sep=',', encoding='iso-8859-1')
```

```python
enem_sp
```

```python
sorocaba = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Sorocaba']
```

```python
sorocaba.shape
```

```python
idades = sorocaba.IDADE
```

```python
idades
```

```python
idades.value_counts(normalize=True) * 100
```

```python
#freq_abs = Counter(idades).most_common()
freq_abs = Counter(idades)
freq_abs
```

```python
# Outra forma, bem melhor

freq_absoluta = sorocaba.IDADE.value_counts()
freq_absoluta
```

```python
freq_abs = pd.DataFrame.from_dict(freq_abs, orient='index')
freq_abs
```

```python
freq_abs = freq_abs.sort_index(ascending=True)
freq_abs
```

```python
freq_abs.rename_axis('Idades', inplace=True)
freq_abs.rename(columns = {0: 'Freq_abs'}, inplace=True) 
freq_abs
```

```python
freq_relativa = freq_abs / freq_abs.sum()
freq_relativa.rename(columns = {'Freq_abs': 'Freq-relativa'}, inplace=True) 
freq_relativa
```

```python
freq_relativa_perc = freq_relativa * 100
freq_relativa_perc.rename(columns = {'Freq-relativa': 'Freq-relativa-per'}, inplace=True) 
freq_relativa_perc
```

## Histograma

### Matplotlib

```python
sorocaba.IDADE.hist(grid=False, bins=60, legend=True, figsize=(12, 5))
plt.title('Distribuição de Idade - Enem SP 2019, Sorocaba')
plt.show()
```

##### Corrigindo erros nas notas

```python
dd.loc[:, 'NU_NOTA_CN'] /= 10
dd.loc[:, 'NU_NOTA_CH'] /= 10
dd.loc[:, 'NU_NOTA_LC'] /= 10
dd.loc[:, 'NU_NOTA_MT'] /= 10
```

```python
dd.head()
```

### Renomear variáveis (colunas)

```python
dd = dd.rename(columns={'NU_NOTA_REDACAO': 'NOTA_REDACAO',
                       'NU_NOTA_CN':'NOTA_CN', 'NU_NOTA_CH':'NOTA_CH',
                       'NU_NOTA_LC': 'NOTA_LC', 'NU_NOTA_MT': 'NOTA_MT'})
```

```python
dd = dd.rename(columns={'NU_NOTA_COMP1': 'COMP1',
                       'NU_NOTA_COMP2':'COMP2', 'NU_NOTA_COMP3':'COMP3',
                       'NU_NOTA_COMP4': 'COMP4', 'NU_NOTA_COMP5': 'COMP5'})
```

```python
dd = dd.rename(columns={'NU_IDADE': 'IDADE',
                       'TP_SEXO':'SEXO', 'TP_COR_RACA':'RACA',
                       'Q025': 'INTERNET', 'TP_ESCOLA': 'ESCOLA'})
```

### Renomeando registros (Linhas)

```python
dd.RACA = dd.RACA.replace({0:'não_declarado', 1: 'branca', 2:'preta', 3:'parda', 4:'amarela',5:'indígena'})
```

```python
dd.INTERNET = dd.INTERNET.replace({'A':'nao', 'B': 'sim'})
```

```python
dd.TP_LINGUA = dd.TP_LINGUA.replace({0:'ingles', 2:'espanhol'})
```

```python
dd.ESCOLA = dd.ESCOLA.replace({1: 'nao_respondeu', 2:'publica', 3: 'privada', 4: 'exterior'})
```
